Lets consider 2 Random variables, X and Y. X can assume values 0 and 1, whereas Y can take values 1, 2, 3. Lets say we perform an experiment 20 times where we check the values of X and Y.

The table above gives the frequency of the event X assumes a value x and Y assumes a value y. For example N(X=0, Y=2) = 5, i.e. the frequency of the outcomes where X was 0 and Y was 2 simultaneously is 5.

Now, lets say you want to calculate the probability of the event where X = 0 and Y = 2 simultaneously would be N(X=0, Y=2)/ Total Frequency = 5/20 = 0.25. This is the joint probability at X = 0 and Y = 2. This is denoted by P(X=0, Y=2). Likewise for any respective values x and y that X and Y can assume simultaneously, will give you the General Joint Probability, P(X=x, Y=y) = P(x, y). Whereas, the table above is the Joint Probability Distribution of the frequencies of the events.

Now, lets say you dont care about whatever value Y takes and you interested in only the number of outcomes where X = 1. In that case you can sum along the 2nd row. There are 3 + 5 + 2 = 10 such outcomes, So the Probability of X = 1 irrespective of the value of Y is P(X=1), which is 10 / 20 = 0.5.
This is the Marginal Probability of X = 1. To get P(X=0) you can sum along the 1st row and divide by the total outcomes, which is again (1 + 5 + 4) / 20 = 10 / 20 = 0.5, i.e. P(X=0) = 0.5. Likewise, to get Marginal Probabilities of Y = 1, 2, 3, you sum along the columns and divide by the total outcomes.

P(Y=1) = 4 / 20 = 0.2; P(Y=2) = 10 / 20 = 0.5; P(Y=3) = 6 / 20 = 0.3
