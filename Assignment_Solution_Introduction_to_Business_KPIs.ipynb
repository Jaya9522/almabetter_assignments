# **1.	What is Customer Retention?**


Customer retention refers to the activities and actions companies and organizations take to reduce the number of customer defections. The goal of customer retention programs is to help companies retain as many customers as possible, often through customer loyalty and brand loyalty initiatives. It is important to remember that customer retention begins with the first contact a customer has with a company and continues throughout the entire lifetime of the relationship.


# **2.	Why is Customer Retention Important?**

* Customer retention is vital to your company’s success. For starters, it is cheaper to cultivate loyalty than to acquire new customers. According to Harvard Business Review, attracting new customers will cost your company five to 25 times more than keeping an existing customer. And loyal customers are likely to be better for your bottom line as well. Inc. reports that current customers spend 67% more on average than those who are new to your business.

* 	Software as a service (or SaaS) is a way of delivering applications over the Internet—as a service. Instead of installing and maintaining software, you simply access it via the Internet, freeing yourself from complex software and hardware management.

* 	In order for a SaaS business to be successful, their customers must be successful first. Acquiring and retaining customers is the most important part of any SaaS business because they rely on monthly recurring revenue (MRR) to grow.

* 	Acquiring new customers can be expensive for SaaS companies, which is why customer retention is so critical. Taking the time to onboard and train new customers can help minimize churn and save time and money for the business.


# **3.	How to track Net Retention?**

* 	Customer retention is tracked by looking at the ratio of Expansion (positive growth from the existing customer base) : Churn (where the whole account is cancelled) & Reduction (where the account downgrades their plan with less MRR).

* 	By calculating net retention on a monthly basis, you’re able to understand how successful your customers are with your product. The goal of any business should be to maintain a net retention of 100% or more. This means that you are bringing in more than you are losing each month, which indicates a monthly growth in your business.


# **4.	What are the relevant KPIs and metrics?**

### **1. Net Revenue Retention Rate**

image.png

**Definition:**

Net Revenue Retention (NRR) Rate is the percentage of recurring revenue retained from existing customers in a defined time period, including expansion revenue, downgrades, and cancels. This churn metric gives a comprehensive view of positive as well as negative changes with respect to customer retention.


**Formula:**

* Sum(RR (recurring revenue) at the beginning of the period + expansion RR during the period - downgraded RR during the period - cancelled RR during the period) / (RR at the beginning of the period)

* Sum(renewing customers RR) / Sum(RR of customers due to renew)


**Key Terms:**

**Cancellations:** Number of customers who have ended their contract with you.

image.png
 
**Contractions:** Customers who have moved to a lower tier or level of your product/service offering.

image.png
 
**Expansions:** Customers who have moved up to a higher tier or level of your product/service offering or have added extra features to their current plan.
 


image.png

**Success Indicator:** 

In a SaaS business, a Net Revenue Retention Rate >100% is a growth indicator. Across all SaaS companies, the median Net Retention Rate is ~100%. Higher Annual Contract Value (ACV) products have a higher Net Retention Rates. For SaaS companies selling into small and medium businesses (SMBs), a good Net Retention Rate is 90%. For Enterprise SaaS, 125% is considered a good Net Retention Rate.


**Example:**

A company has 100 customers, each paying Rs. 2,000 per month. MRR at the beginning of the month is Rs. 200,000. Within the month, 1 customer adds a Rs. 4,000 MRR upgrade, 2 downgrade by Rs. 500 each, and 1 customer cancels. NRR  = (Rs. 200,000 + Rs. 4,000 - (Rs. 500 x 2) - Rs. 2,000) / Rs. 200,000 = Rs. 201,000 / Rs. 200,000 = 100.5% expressed monthly


### **2. Monthly Recurring Revenue (MRR)**

image.png

image.png

**Definition:**

For subscription based businesses like SaaS companies, telecommunication companies and banks, monthly recurring revenue is at the center of the business model, and is one of the most important KPIs they monitor.
Monthly Recurring Revenue targets can vary drastically from industry to industry, but should reflect a company’s position in its lifecycle, its growth objectives and growth timelines.


**Formula:**

Number of subscribers under a monthly plan * average revenue per user (ARPU)


**Key Term:**

**Cancelled Accounts:** Accounts who have cancelled their subscription with your product/service.

image.png
 



**Example:**

For instance, suppose you have 5 subscribers on the Rs. 300/month plan. The MRR will be:
(5* Rs. 300) = Rs. 1500

For subscriptions under annual plans, MRR is calculated by dividing the annual plan price by 12 and then multiplying the result by the number of customers on the annual plan.

For instance, if a customer subscribes to your product with a monthly renewal agreement for Rs. 500 per month, the ARR would be, Rs. 500 * 12 = Rs. 6000


### **3. Customer Acquisition Cost (CAC)**

image.png

**Definition:**

Customer Acquisition Cost (CAC) measures the costs of acquiring new customers by adding up sales and marketing costs for a given period, and dividing them by the number of new customers for that period. CAC calculations can vary in regards to the inclusion of variable and fixed (or overhead) costs. 


**Formula:** 

Total sales costs + total marketing costs for a period / number of new customers over that period


**Example:**

Consider a SaaS company that spent Rs. 125,000 on sales and marketing in a month, including salaries, commissions, ad spend and trial support. 50 new customers were signed up during the same month.

In this case:

CAC = Rs. 125,000/50
= Rs. 2,500


### **4. Customer Lifetime Value to Customer Acquisition Cost (LTV:CAC)**

image.png

**Defintion:**

The Customer Lifetime Value to Customer Acquisition (LTV:CAC) ratio measures the relationship between the lifetime value of a customer, and the cost of acquiring that customer. The metric is computed by dividing LTV by CAC. It is a signal of customer profitability, and of sales and marketing efficiency.


**Formula:**

Customer Lifetime Value / Customer Acquisition Cost


**Example:**

Consider a SaaS company that had a gross margin of 75% and monthly customer churn of 2%, and each customer spent an average of Rs. 40 with you every month, the calculation would look like this:
75% X ( 1 / 2% ) X Rs. 40 = Rs. 1,500 LTV


### **5.	Customer Retention Rate**

image.png

**Definition:**

Customer Retention Rate calculates the proportion of customers in a particular period that are retained into the next period.


**Formula:**

*  (Count(Customers End of Period) - Count(Acquired Customers)) / Count(Customers Beginning of Period)


**Example:**

A business had 1,100 customers at the beginning of the year and has 1,000 customers at the end of the year, out of which 100 are newly acquired. In this case, the Customer Retention Rate would be (1,000 - 100) / 1,100 which is approximately 81%.
